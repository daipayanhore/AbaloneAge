```python
#Loading important packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading csv file
data= pd.read_csv("abalone.data.csv")
data
```

```python
#Checking the data graphically
x= data['Rings']
y= data['Age']
plt.scatter(x,y)
plt.xlabel("Rings")
plt.ylabel("Age")
plt.title("Age v. Rings")
plt.show()
```

```python
#Creating a dataframe with just desired attributes
df= data[['Rings', 'Age']]
df.columns = ['Rings', 'Age']
df
```

```python
df.info()   #Checking the information of selected attributes
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr

x= np.array(x).reshape(-1,1)
y= np.array(y).reshape(-1,1)
x_train, x_test, y_train, y_test = tts(x,y,test_size = 0.20, random_state= 0)
reg= lr()
reg.fit(x_train, y_train)
y_pred= reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
plt.scatter(x,y,s= 5, c= 'green')  #s for size, c for color
plt.plot(x_test, y_pred,c= 'red', linewidth= 2)
plt.xlabel("Number of Rings", color= "blue")
plt.ylabel("Age", color ="blue")
plt.title("Linear Regression Model version UNO", size= 20)
plt.show()
```
